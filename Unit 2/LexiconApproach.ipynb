```python
# SentimentIntensityAnalzer - tool which uses Lexicon Approach ( to determine sentiment of text)
import nltk
import spacy
import re # regular expression
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
# Vader- List of words with sentiment scores
nltk.download('vader_lexicon')
nlp = spacy.load("en_core_web_sm")
```

```python
# Initialize the analyzer
sia = SentimentIntensityAnalyzer()
```

```python
# Sample review text
review = "The product is amazing! & works perfectly."
```

```python
# re.sub(pattern, repl, string)
# search for substitute
# match anything that is not a word character or whitespace
cleaned_review = re.sub(r'[^\w\s]', '', review)
```

```python
doc = nlp(cleaned_review)
```

```python
print(f"Original Review: {review}")
print(f"Cleaned Review: {cleaned_review}")
```
