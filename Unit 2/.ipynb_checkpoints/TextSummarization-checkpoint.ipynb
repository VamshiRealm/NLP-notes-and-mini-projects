```python
# parse the plain text text -- structured format
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.text_rank import TextRankSummarizer
# helps to select the most imp sentences from the text
```

```python
# Sample text for summarization
text = """
Text summarization is the process of shortening a set of data computationally. 
It is one of the most crucial parts of natural language processing (NLP) and is widely used in many applications 
like search engines and news aggregators. The summarization can be done in two main ways: extractive and abstractive.
Extractive summarization extracts key sentences or phrases from the original text, while abstractive summarization 
generates new sentences that convey the most critical information. In this tutorial, we will focus on extractive summarization 
using the TextRank algorithm.
"""
```

```python
# Step 1: Process the text and parse it
parser = PlaintextParser.from_string(text, Tokenizer('english')) #specify the lang
```

```python
# Step 2: Initialize TextRank Summarizer
summarizer = TextRankSummarizer()
```

```python
# Step 3: Generate the summary by extracting top 2 sentences
summary = summarizer(parser.document, 2)
```

```python
# Step 4: Print the summary
for sentence in summary:
    print(sentence)
```
