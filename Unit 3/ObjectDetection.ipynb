```python
# Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train
```

```python
# Normalize the images
x_train, x_test = (x_train/255.0), (x_test/255.0)
x_train
```

```python
# label for categorical values-converted into integer
print(f"Total Categories: {y_train.shape}")
print(y_train[:5])
```

```python
# One-hot encoding the labels
ytrain_onehot = to_categorical(y_train, 10)
ytest_onehot = to_categorical(y_test, 10)
ytrain_onehot
```

```python
# Split the training data into train and validation sets
x_train, x_val, ytrain_onehot, yval_onehot = train_test_split(x_train, ytrain_onehot, random_state = 42, test_size = 0.2)
```

```python
# Build the CNN model
model = Sequential()
```

```python
# 1st convolution layer
model.add(Conv2D(32, (3,3), activation = "relu", input_shpae=(32, 32,3)))
model.add(MaxPooling2D(pool_size = (2,2)))
```

```python
# 2nd Convolution layer
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooloing2D(poolsize=(2,2)))
```

```python
# 3rd Conv layer
model.add(Conv2D(128, (3,3), activation='relu'))
model.add(MaxPooloing2D(poolsize=(2,2)))
```

```python
# Flattening the layers
model.add(Flatten())

# Fully Connected layer
model.add(Dense(128, activation='relu'))

# Output layer
model.layer(Dense(10, activation='softmax'))
```

```python
model.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history = model.fir(x_train, ytrain_onehot, validation_data = (x_val, yval_onehot), epochs = 5, batch_size = 64)
```

```python
# Evaluate the model on test data
test_loss, test_accuracy = model.evaluate(x_test, ytest_onehot)

print(f"Test Accuracy: {test_accuracy*100: .2f}%")
```
