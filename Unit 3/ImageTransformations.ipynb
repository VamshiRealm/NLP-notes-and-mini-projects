```python
import cv2
import numpy as np

# Load an image
image = cv2.imread(r"C:\Users\vamsh\Pictures\party_ledha _pushpa.jpg")
image
# [r, g, b] output
```

```python
image.shape
```

```python
# Translation -output will be numpy array format
# scaling rotation translation
M = np.float32([[1,0,50], [0,1,30]])
translated_image = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
translated_image
```

```python
# Scaling
scaled_image = cv2.resize(image, None, fx=2, fy=2)
```

```python
# Affine transformation (rotation + translation)
rows, cols , _ = image.shape
M = cv2.getRotationMatrix2D((cols/2, rows/2), 45, 1)
tx = 50 # translation in x-axis
ty = 30 #translation in y-axis
M[0,2] += tx
M[1,2] += ty
affine_image = cv2.warpAffine(image, M, (cols, rows))
```

```python
# Shearing along the x-axis
shear_x = np.float32([[1,0.5, 0], [0,1,0]]) #Shear-factor of 0.5 along x-axis
sheared_image_x = cv2.warpAffine(image, shear_x, (image.shape[1], image.shape[0]))

shear_y = np.float32([[1,0, 0], [0.5,1,0]]) #Shear-factor of 0.5 along y-axis
sheared_image_y = cv2.warpAffine(image, shear_y, (image.shape[1], image.shape[0]))
```

```python
# Show Images
cv2.imshow("Translated", translated_image)
cv2.imshow("Scaled", scaled_image)
cv2.imshow("Affined: rotation + translation", affine_image)
cv2.imshow("Sheared Along x:", sheared_image_x)
cv2.imshow("Sheared Along y:", sheared_image_y)


key = cv2.waitKey(0)
if key == 27: #ASCII value of 'esc' is 27
    print("Escape key pressed, closing window...")
else:
    print(f"Key pressed: {key}")

cv2.destroyAllWindows()
# closes all windows when program finishes
```
