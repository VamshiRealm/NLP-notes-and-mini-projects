```python
import tensorflow as tf
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Load and preprocess an image
def preprocess_image(img_path, target_size=(128,128)):
    img = Image.open(img_path).resize(target_size)
    return np.expand_dims(np.array(img) / 255.0, axis=0)
```

```python
#simple U-net model
def simple_unet(input_shape=(512,512,3)):
    
    #defines the input layer that takes images of given shape
    inputs=tf.keras.layers.Input(input_shape)
    
    #Add the convlution layer ehich applies 64 filter of size 3x3
    #keeps the output size the same as the input
    x=tf.keras.layers.Conv2D(64,3,activation='relu',padding='same')(inputs)
    
    #Transposed convolution:upsamples the features map to restore spatial 
    x=tf.keras.layers.Conv2DTranspose(64,3,activation='relu',padding='same')(x)
    
    #Final output layer:uses a 1x1 convolution to generate a single-chaneel output
    outputs=tf.keras.layers.Conv2D(1,1,activation='sigmoid')(x)
    return tf.keras.models.Model(inputs,outputs)

model=simple_unet()
img_path=r"C:\Users\vamsh\Pictures\IMG_2818.JPG"
img_array=preprocess_image(img_path, target_size=(512,512))
prediction = (model.predict(img_array)[0, : , : , 0] > 0.5).astype(np.uint8)
plt.imshow(prediction, cmap="gray")
plt.axis("off")
plt.show()
```
