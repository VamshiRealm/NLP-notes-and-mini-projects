```python
import cv2
import numpy as np

# Read img and cnvt into gray scale
image = cv2.imread(r"C:\Users\vamsh\Pictures\IMG_2818.JPG", cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image, (512,512))
```

```python
# Apply threshold
_, thresh = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Find contours
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw the contours
output = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
cv2.drawContours(output, contours, -1, (0,255,0), 2)

cv2.imshow('Contours', output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# load the image
image = cv2.imread(r"C:\Users\vamsh\Pictures\IMG_2818.JPG", cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image, (512,512))
```

```python
# Creates a black mask ( all pixels set to 0)
mask = np.zeros(image.shape[:2], dtype=np.uint8)

# Define a circular ROI (250-senter of circle, 100- radius, 255- white
cv2.circle(mask , (250,250), 100, 255, -1) #white circle

# Apply the mask to the image - which pixels to keep
masked_image = cv2.bitwise_and(image, image, mask = mask)

cv2.imshow('Original Image', image)
cv2.imshow('Mask', mask)
cv2.imshow('Masked Image', masked_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
